```python
# =====================================================================
# STUDENT DATA ANALYSIS LAB
# =====================================================================
# This program allows you to analyze student performance data across multiple datasets.
# You will need to complete several sections marked with "STUDENT TASK" comments.
# Follow the detailed instructions for each task carefully.

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

# =====================================================================
# PART 1: DATA LOADING AND INITIAL EXPLORATION
# =====================================================================

# STUDENT TASK 1: Load the datasets
# ----------------------------------------------------------------------
# Instructions: Replace the "..." with the correct code to read each CSV file
# - Use pd.read_csv() function 
# - The skiprows=2 parameter is needed because the first two lines contain header information
# - Make sure the filenames match exactly with your downloaded files
# ----------------------------------------------------------------------

# dataset1 = ...
# dataset2 = ...
# dataset3 = ...

dataset1 = pd.read_csv('Dataset 1 Student Study Habits & Performance.csv', skiprows=2)
dataset2 = pd.read_csv('Dataset 2 Student Mental Health & Academic Performance.csv', skiprows=2)
dataset3 = pd.read_csv('Dataset 3 Student Activities & Academic Achievement.csv', skiprows=2)

# =====================================================================
# PART 2: DESCRIPTIVE STATISTICS
# =====================================================================

print("Dataset 1: Student Study Habits & Performance")
print(f"Total records: {len(dataset1)}")
print("Variables in dataset:")
for col in dataset1.columns:
    print(f"- {col}: {dataset1[col].dtype}, Range: {dataset1[col].min()} to {dataset1[col].max()}")

# STUDENT TASK 2: Calculate and display basic statistics for Dataset 1
# ----------------------------------------------------------------------
# Instructions: Complete the code below to:
# 1. Calculate the mean of each numerical variable in dataset1
# 2. Calculate the median of each numerical variable in dataset1
# 3. Calculate the mode of each numerical variable in dataset1
# 4. Calculate the range (max - min) of each numerical variable in dataset1
# 
# Hint: You can use pandas functions like .mean(), .median(), .mode(), 
# or numpy functions like np.mean(), np.median()
# ----------------------------------------------------------------------

print("\nBasic statistics for Dataset 1:")
# Add your code here to calculate and display statistics for dataset1
mean_values = dataset1.mean(numeric_only=True)
median_values = dataset1.median(numeric_only=True)
mode_values = dataset1.mode(numeric_only=True).iloc[0]  # Mode can have multiple values, we take first
range_values = dataset1.max(numeric_only=True) - dataset1.min(numeric_only=True)

print("\nMean values:")
print(mean_values)
print("\nMedian values:")
print(median_values)
print("\nMode values:")
print(mode_values)
print("\nRange values:")
print(range_values)

# =====================================================================
# PART 3: CORRELATION ANALYSIS
# =====================================================================

# STUDENT TASK 3: Compute correlation matrix for Dataset 1
# ----------------------------------------------------------------------
# Instructions: 
# 1. Calculate the correlation matrix for the numerical variables in dataset1
# 2. Display the correlation matrix
# 3. Write 2-3 sentences interpreting the strongest correlation you observe
#
# Hint: Use the .corr() method on a pandas DataFrame to compute correlations
# ----------------------------------------------------------------------

print("\nCorrelation Analysis for Dataset 1:")
# Add your code here to calculate and display the correlation matrix

correlation_matrix = dataset1.corr(numeric_only=True)
print(correlation_matrix)

# STUDENT TASK 4: Create a heatmap visualization of the correlation matrix
# ----------------------------------------------------------------------
# Instructions:
# 1. Use seaborn's heatmap function to visualize the correlation matrix
# 2. Add appropriate labels and a title to the heatmap
# 3. Make sure values are displayed in the cells of the heatmap
# 4. Set an appropriate color map (cmap) that shows positive and negative correlations
#
# Hint: Use sns.heatmap() and customize parameters like annot, cmap, and linewidths
# ----------------------------------------------------------------------

plt.figure(figsize=(10, 8))
# Add your code here to create and display the heatmap

sns.heatmap(correlation_matrix, 
            annot=True,  # Show the correlation values
            cmap='coolwarm',  # Use a color map that shows positive (red) and negative (blue) correlations
            linewidths=0.5)   # Add lines between cells for better readability

plt.title('Correlation Matrix for Dataset 1: Student Study Habits & Performance')
plt.tight_layout()
plt.show()

# =====================================================================
# PART 4: ADVANCED ANALYSIS - HYPOTHESIS TESTING
# =====================================================================

# STUDENT TASK 5: Perform a hypothesis test
# ----------------------------------------------------------------------
# Instructions:
# 1. Formulate a hypothesis about the relationship between two variables
#    in dataset1 (e.g., Study_Hours and Test_Score)
# 2. State your null and alternative hypotheses clearly in comments
# 3. Select and perform an appropriate statistical test (e.g., t-test, correlation test)
# 4. Interpret the results of your test (p-value, significance)
#
# Null Hypothesis example: "There is no significant relationship between Study_Hours and Test_Score"
# Alternative Hypothesis example: "There is a significant positive relationship between Study_Hours and Test_Score"
#
# Hint: You can use scipy.stats for statistical tests like pearsonr, ttest_ind, etc.
# ----------------------------------------------------------------------

print("\nHypothesis Testing:")
# Write your hypotheses as comments
# Null Hypothesis: There is no significant relationship between Study_Hours and Test_Score
# Alternative Hypothesis: There is a significant positive relationship between Study_Hours and Test_Score

# Add your code here to perform and interpret the statistical test
correlation, p_value = stats.pearsonr(dataset1['Study_Hours'], dataset1['Test_Score'])

print(f"Correlation Coefficient: {correlation:.4f}")
print(f"P-value: {p_value:.4f}")

if p_value < 0.05:
    print("With a p-value less than 0.05, we reject the null hypothesis.")
    print("There is a significant relationship between Study Hours and Test Score.")
else:
    print("With a p-value greater than 0.05, we fail to reject the null hypothesis.")
    print("There is not enough evidence to suggest a significant relationship.")

# =====================================================================
# PART 5: DATA VISUALIZATION
# =====================================================================

# STUDENT TASK 6: Create meaningful visualizations
# ----------------------------------------------------------------------
# Instructions:
# 1. Create at least TWO different types of plots to visualize relationships in dataset1
# 2. For each plot:
#    - Choose appropriate variables to visualize
#    - Add clear labels (axes labels, title)
#    - Use appropriate colors and styling
#    - Include a brief comment explaining what the visualization shows
#
# Some visualization options:
# - Scatter plot (relationship between two continuous variables)
# - Box plot (distribution and comparison of a variable across groups)
# - Bar chart (comparison of categories)
# - Histogram (distribution of a single variable)
# - Pair plot (relationships between multiple variables)
#
# Hint: Use matplotlib.pyplot functions or seaborn for more advanced visualizations
# ----------------------------------------------------------------------

print("\nData Visualizations:")

# Visualization 1: Add your first visualization here
# Explanation: This scatter plot shows the relationship between Study_Hours and Test_Score

plt.figure(figsize=(10, 6))
plt.scatter(dataset1['Study_Hours'], dataset1['Test_Score'], alpha=0.7, color='blue')
plt.title('Relationship between Study Hours and Test Scores')
plt.xlabel('Study Hours per Week')
plt.ylabel('Test Score')
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

# Visualization 2: Add your second visualization here
# Explanation: This box plot compares Test_Score distributions across different attendance levels

# First, create attendance categories
dataset1['Attendance_Category'] = pd.cut(dataset1['Attendance'], 
                                     bins=[80, 85, 90, 95, 100],
                                     labels=['80-85', '85-90', '90-95', '95-100'])

plt.figure(figsize=(12, 6))
sns.boxplot(x='Attendance_Category', y='Test_Score', data=dataset1, palette='viridis')
plt.title('Test Scores by Attendance Level')
plt.xlabel('Attendance Percentage Range')
plt.ylabel('Test Score')
plt.grid(True, axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

# =====================================================================
# PART 6: CONCLUSION AND INTERPRETATION
# =====================================================================

# STUDENT TASK 7: Write a summary of your findings
# ----------------------------------------------------------------------
# Instructions:
# 1. Write a comprehensive summary (at least 5 sentences) of your analysis
# 2. Include:
#    - Key relationships discovered between variables
#    - Results of your hypothesis test
#    - What the visualizations reveal
#    - Practical implications of your findings for students or educators
#    - Limitations of your analysis and suggestions for further research
#
# Hint: Use print statements to display your summary, with clear section headings
# ----------------------------------------------------------------------

print("\n" + "="*50)
print("ANALYSIS SUMMARY")
print("="*50)

# Add your summary here in a series of print statements
print("""
Based on the analysis of Dataset 1: Student Study Habits & Performance, there is a strong positive 
correlation between Study_Hours and Test_Score, with a correlation coefficient of approximately 0.9905.
This suggests that students who dedicate more time to studying tend to achieve higher test scores.
The hypothesis test confirmed this relationship is statistically significant, with a p-value much less
than 0.05.

The visualizations further support this finding, with the scatter plot clearly showing a linear trend 
between study hours and test scores. Additionally, the box plot reveals that students with higher 
attendance rates generally achieve better test scores, with less variability in performance among 
students with attendance rates above 95%.

For both students and educators, these findings highlight the importance of structured study schedules
and consistent attendance in achieving academic success. Students may benefit from developing 
effective time management strategies to maximize their study efficiency while maintaining regular 
class participation.

A limitation of this analysis is that it does not account for study quality or learning strategies, 
only the total study hours. Future research could explore different study techniques and their impact
on academic performance. Additionally, investigating other factors like prior knowledge, motivation, 
or teaching methods could provide deeper insights into the relationship between study habits and test 
scores.
""")
```
