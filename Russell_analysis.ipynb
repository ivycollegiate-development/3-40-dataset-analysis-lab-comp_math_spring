```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

# Load the datasets
dataset1 = pd.read_csv('Dataset 1 Student Study Habits & Performance.csv', skiprows=2)
dataset2 = pd.read_csv('Dataset 2 Student Mental Health & Academic Performance.csv', skiprows=2)
dataset3 = pd.read_csv('Dataset 3 Student Activities & Academic Achievement.csv', skiprows=2)

# Display basic information about each dataset
print("Dataset 1: Student Study Habits & Performance")
print(f"Total records: {len(dataset1)}")
print("Variables in dataset:")
for col in dataset1.columns:
    print(f"- {col}: {dataset1[col].dtype}, Range: {dataset1[col].min()} to {dataset1[col].max()}")
print("\nBasic statistics for Dataset 1:")
print(dataset1.describe())

print("\n\nDataset 2: Student Mental Health & Academic Performance")
print(f"Total records: {len(dataset2)}")
print("Variables in dataset:")
for col in dataset2.columns:
    print(f"- {col}: {dataset2[col].dtype}, Range: {dataset2[col].min()} to {dataset2[col].max()}")
print("\nBasic statistics for Dataset 2:")
print(dataset2.describe())

print("\n\nDataset 3: Student Activities & Academic Achievement")
print(f"Total records: {len(dataset3)}")
print("Variables in dataset:")
for col in dataset3.columns:
    print(f"- {col}: {dataset3[col].dtype}, Range: {dataset3[col].min()} to {dataset3[col].max()}")
print("\nBasic statistics for Dataset 3:")
print(dataset3.describe())
```
